# Chemin projet

```python
from pathlib import Path

PROJECT_DIR = Path.cwd().parent  # notebooks/ -> projet racine
DATA_DIR = PROJECT_DIR / "data"
RAW_DIR = DATA_DIR / "raw"
INTERIM_DIR = DATA_DIR / "interim"
PROCESSED_DIR = DATA_DIR / "processed"
REPORTS_DIR = PROJECT_DIR / "reports"

print("PROJECT_DIR :", PROJECT_DIR)
print("RAW_DIR     :", RAW_DIR)
print("INTERIM_DIR :", INTERIM_DIR)
print("PROCESSED   :", PROCESSED_DIR)
print("REPORTS_DIR :", REPORTS_DIR)

# Création au besoin (sécurise le projet)
for p in [RAW_DIR, INTERIM_DIR, PROCESSED_DIR, REPORTS_DIR / "figures", REPORTS_DIR / "outputs"]:
    p.mkdir(parents=True, exist_ok=True)

print("\n Dossiers vérifiés/créés.")
```
